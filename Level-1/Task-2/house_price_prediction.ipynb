```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("house Prediction Data Set.csv",header=None,sep=r'\s+')
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.columns = column_names
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.corr()['MEDV'].sort_values(ascending=False)
```

```python
X = df[['RM']]
Y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
df['RM'].hist(bins=30)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
lre = LinearRegression()
lre.fit(x_train,y_train)
```

```python
y_pred = lre.predict(x_test)
```

```python
y_pred
```

```python
lre.coef_
```

```python
from sklearn.metrics import mean_squared_error,r2_score

y_train_ped = lre.predict(x_train)
y_test_ped = y_pred
```

```python
mse_train = mean_squared_error(y_train,y_train_ped)
mse_test = mean_squared_error(y_test,y_test_ped)
print(mse_train,mse_test)
```

```python
score_train = r2_score(y_train,y_train_ped)
score_test = r2_score(y_test,y_test_ped)
print(score_train,score_test)
```
